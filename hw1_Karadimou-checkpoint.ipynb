```python
def parsingout():
    Id  = []
    CreationDate = []
    OwnerUserId = []
    FirstTag = []
    file = 'stackoverflow-posts-2015.xml' #to be tested in the same directory
    par = ET.iterparse(file, events = ('end','start'))
    for ev, el in par:
        if ev == 'end' and el.tag == 'row':
            pTypeId = el.get('PostTypeId')
            if pTypeId == '1':
                Id.append(el.get('Id'))
                CreationDate.append(el.get('CreationDate'))
                OwnerUserId.append(el.get('OwnerUserId'))
                FirstTag.append(el.get('Tags').split("<")[1].split(">")[0].strip()) #splitting between <> to take the first tag
                
                
    data = {"Id": Id,
           "CreationDate" : CreationDate,
           "OwnerUserId" : OwnerUserId,
           "FirstTag" : FirstTag}
    frame = pd.DataFrame(data)
    print(frame)
    return(frame)
```

```python
def parsingout2(): #for both types of posts (questions and answers)
    Id  = []
    IdA = []
    CreationDate = []
    OwnerUserId = []
    OwnerUserIdA = []
    FirstTag = []
    ParentId = []
    file = 'stackoverflow-posts-2015.xml' 
    par = ET.iterparse(file, events = ('end','start'))
    for ev, el in par:
        if ev == 'end' and el.tag == 'row':
            pTypeId = el.get('PostTypeId')
            if pTypeId == '1':
                Id.append(el.get('Id'))
                CreationDate.append(el.get('CreationDate'))
                OwnerUserId.append(el.get('OwnerUserId'))
                FirstTag.append(el.get('Tags').split("<")[1].split(">")[0].strip()) 
            if pTypeId == '2':
                ParentId.append(el.get('ParentId'))
                OwnerUserIdA.append(el.get('OwnerUserId'))
                IdA.append(el.get('Id'))
                
                
                
    data = {"Id": Id,
           "CreationDate" : CreationDate,
           "OwnerUserId" : OwnerUserId,
           "FirstTag" : FirstTag}
    ansdata = {"Id": IdA,
              "OwnerUserId" : OwnerUserIdA,
               "ParentId": ParentId}
    frame = pd.DataFrame(data)
    ansframe = pd.DataFrame(ansdata)
    print(frame)
    print(ansframe)
    return(ansframe)
ansframe = pd.DataFrame(parsingout2())
frame = pd.DataFrame(parsingout())
```

```python
def getTop100(frame,ansframe):
    uId_ParentId = ansframe[['OwnerUserId','ParentId']]
    top = ansframe.OwnerUserId.value_counts().index.values[:100]
    usrs = {}
    for id in top:         
        tags = frame.FirstTag[frame.Id.isin(uId_ParentId[uId_ParentId.OwnerUserId == id].ParentId)]
        usrs[id] = set(tags)
    print(top)
    return (usrs, top)
```

```python
getTop100(frame,ansframe)
```

```python
import numpy as np

#will use Jaccard distance matrix to compute similarity based on tags
def distance(usrs, top):
    matrix = np.zeros((100,100))
    i = -1
    for us1 in top:
        i = i + 1
        j = -1
        if us1 in usrs:
            set1 = usrs[us1]
            for us2 in top:
                j = j + 1 
                if us2 in usrs:
                    set2 = usrs[us2]
                    distanceJ = 1 - (float(len(set1.intersection(set2))) / len(set1.union(set2)))
                    matrix[i][j] = distanceJ
    return matrix
```

```python
( D, top) = getTop100(frame, ansframe)
```

```python
matrix = distance(D, top)
```
